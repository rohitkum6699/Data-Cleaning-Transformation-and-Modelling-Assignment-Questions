
## Question 1: What is Data Loading in Power BI and why is it considered the first step of analysis?

**Data Loading** in **Power BI** is the process of **importing data from different data sources into Power BI so that it can be analyzed and visualized**.

It involves connecting to sources such as:

* Excel files
* CSV files
* Databases (MySQL, SQL Server)
* Web APIs
* Cloud services

For example, in this dataset you would load the **COVID-19 dataset CSV file** into Power BI.

**Why it is the first step of analysis:**

1. Power BI cannot perform analysis without data.
2. Data must be imported before cleaning, transforming, and modeling.
3. It allows Power BI to understand the **structure of the dataset (columns and rows)**.
4. After loading, we can build **relationships, measures, and dashboards**.

Example from dataset after loading:

* Country wise **Confirmed Cases**
* Date wise **Deaths**
* State wise **Hospitalization**

So, **data loading is the foundation of the entire data analysis process in Power BI.**








## Question 2: Explain the difference between “Load” and “Transform Data” in Power BI.

| Feature       | Load                                    | Transform Data                                        |
| ------------- | --------------------------------------- | ----------------------------------------------------- |
| Meaning       | Directly imports the data into Power BI | Opens **Power Query Editor** to clean and modify data |
| Data Cleaning | No cleaning is performed                | Data cleaning and transformation are performed        |
| Usage         | Used when data is already clean         | Used when data needs preparation                      |
| Tool Used     | Power BI Model                          | Power Query                                           |

**Example using this dataset:**

If we click **Load**:

* The dataset is imported directly with columns like *Country, State, Confirmed_Cases*.

If we click **Transform Data**:
We can perform operations like:

* Remove **null values in Vaccination_Status**
* Convert **Date format**
* Remove duplicates
* Rename columns

So:

* **Load → Direct import**
* **Transform Data → Data cleaning and preparation**

## Question 3: What is a Fact Table and a Dimension Table? Give examples from the dataset.

### Fact Table

A **Fact Table** contains **numerical and measurable data used for analysis**.

Characteristics:

* Contains **metrics**
* Usually large
* Linked to dimension tables through keys

**Example from this dataset (Fact Table):**

COVID Statistics Table

Measures:

* Confirmed_Cases
* Recovered_Cases
* Deaths
* Active_Cases
* Tests_Conducted
* Hospitalized

These are **quantitative values used for analysis**.

Example record:

| ------- | ---- | --------------- | ------ |
| Brazil  | 2020 | 5000            | 120    |

* Used for filtering and grouping

## Question 4: Why is Star Schema preferred over Snowflake Schema in Power BI?

A **Star Schema** is preferred because it is **simpler, faster, and more efficient for reporting and analytics**.

### 1. Better Performance

Star schema has **fewer joins**, so Power BI queries run faster.

### 2. Simple Structure

It has:

* **One Fact Table**
* **Multiple Dimension Tables**

This structure looks like a **star shape**.

Example:

```
        Country
           |
Date ---- Fact Table ---- Vaccination
           |
        State
```

### 3. Easier for Data Analysis

Business users and analysts can easily understand the model.

### 4. Better for Power BI Engine

Power BI uses the **VertiPaq engine**, which works best with **denormalized star schema models**.

### 5. Faster Report Creation

Measures like:

* Total Confirmed Cases
* Death Rate
* Recovery Rate

can be calculated quickly.

---

### Difference Summary

| Feature           | Star Schema | Snowflake Schema |
| ----------------- | ----------- | ---------------- |
| Structure         | Simple      | Complex          |
| Joins             | Fewer       | More             |
| Performance       | Faster      | Slower           |
| Maintenance       | Easy        | Difficult        |
| Best for Power BI | Yes         | Not preferred    |
